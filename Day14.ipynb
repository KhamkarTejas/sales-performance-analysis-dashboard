```python
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\Admin\Desktop\Tejas Khamkar\Practice Material\Datasets\Dataset13.csv")
df
```

```python
region_summary = df.groupby("Region")[['Sales','Target']].sum()

ax = region_summary.plot(kind='bar',figsize =(10,5))

plt.title("Sales vs Target by Region")
plt.xticks(rotation=0)

# Add data labels
for container in ax.containers:
    ax.bar_label(container)

plt.show()
```

```python
df["Date"] = pd.to_datetime(df["Date"], dayfirst= True )
```

```python
df.info()
```

```python
# Create new column for Month Name
df["Month"] = df["Date"].dt.month_name()

# Group sales by Month
monthly_trend = df.groupby("Month")["Sales"].sum()

month_order = ["January","February","March","April","May","June",
               "July","August","September","October","November","December"]

monthly_trend = monthly_trend.reindex(month_order)

# Plot graph
monthly_trend.plot(figsize=(8,4))

plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly Sales Trend")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\Admin\Desktop\Tejas Khamkar\Practice Material\Datasets\Day14_Sales_Dataset.csv")

df.head()
```

```python
df.info()
```

```python
df.columns = df.columns.str.lower().str.replace(" ","_")
df.columns
```

```python
df['order_date'] = pd.to_datetime(df['order_date'], dayfirst=True ,errors = 'coerce')
df.info()
```

```python
df.describe()
```

```python
# Filter high sales records
high_sales_records = df[df['sales_category'] == "High"].sort_values(by="sales", ascending =False )
high_sales_records.head(10)
```

```python
# Monthly Sales Trend chart

df['month'] = df['order_date'].dt.month_name()
monthly_sales = df.groupby("month")['sales'].sum()

month_order = ["January","February","March","April","May","June",
               "July","August","September","October","November","December"]

monthly_sales = monthly_sales.reindex(month_order)

# Plot graph
monthly_sales.plot(figsize=(8,3))

plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly Sales Trend")
plt.show()
```

```python
category_sales = df.groupby("category")['sales'].sum()

# Plot graph
ax = category_sales.plot(kind = "bar",figsize=(8,4))

for container in ax.containers:
    ax.bar_label(container, label_type = "center")

plt.xlabel("Category")
plt.ylabel("Sales")
plt.title("Sales by Category")
plt.xticks(rotation = 0)
plt.show()
```
